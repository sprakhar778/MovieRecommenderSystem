```python
import numpy as np
import pandas as pd
import ast
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
# movies.head()
# movies.merge(credits,on='title').shape
# movies.shape
# credits.shape
movies=movies.merge(credits,on='title')
```

```python
# movies.head(1)
# movies.info()
# movies['original_language'].value_counts()
#genres
#id
#keywords
#title
#overview
#cast   
#crew
movies=movies[['id','genres','keywords','title','overview','cast','crew']]
movies.isnull().sum()
movies.dropna(inplace=True)
movies.isnull().sum()
movies.duplicated().sum()



```

```python
movies.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, 
# {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#List is a string so change it
import ast

movies['genres'][0]
```

```python


#literal_eval(arr) is used to change string into appropriate format.
def convert(arr):
    L=[]
    for i in ast.literal_eval(arr.replace("null", "None")):
#         print(i,i['name'])
        L.append(i['name'])
    return L

# s = '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# print(convert(s))
```

```python


movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# movies['cast'][0]
def convert3(arr):
    L=[]
    c=0
    for i in ast.literal_eval(arr):
        if c<3:
            L.append(i['name'])
            c+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(arr):
    L=[]
    for i in ast.literal_eval(arr):
        if i['job']=='Director':
            L.append(i['name'])
    return L
        
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
#string change it into list
```

```python
movies['overview']=movies['overview'].apply(lambda x : x.split())
```

```python
#'Sam Washington'
#'SamWashington'
#remove the spaces
movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(5)
```

```python
new_df=movies[['id','title','tags']]
new_df.head(5)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head(5)
```

```python
new_df['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
# cv.fit_transform(new_df['tags']).toarray().shape
vectors=cv.fit_transform(new_df['tags']).toarray()
# vectors[0][:1000]
vectors
# len(vectors[0])
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity

# cosine_similarity(vectors).shape
```

```python
def recommend(movies):
#     new_df[new_df['title']=='The Lego Movie'].index[0]
    movies_index=new_df[new_df['title']==movies].index[0]
    vector=similarity[movies_index]
    movies_list=sorted(list(enumerate(vector)),reverse=True,key=lambda x:x[1])[1:6]
    for i in range(5):
#         print(movies_list[i][1])
          print(new_df['title'][movies_list[i][0]])
   
#         print(new_df['title'][movies_list[i][0]] ,"with" , int(movies_list[i][1]*100) ,"% similiarity")
   
```

```python
recommend('Avatar')
```
